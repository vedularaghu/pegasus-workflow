```python
import logging
from pathlib import Path
from Pegasus.api import *

logging.basicConfig(level = logging.DEBUG)

props = Properties()
props["dagman.retry"] = "5"
props.write()

fa = File("saved_stated.txt")
rc = ReplicaCatalog()\
    .add_replica("local", fa, Path(".").resolve() / "saved_stated.txt")\
    .write()

add_nums = Transformation(
                "add_nums",
                site="local",
                pfn=str(Path(".").resolve() / "print_nums.py"),
                is_stageable=True
            )

tc = TransformationCatalog().add_transformations(add_nums).write()

wf = Workflow("add_nums")

job_add_nums = Job(add_nums)\
                .add_checkpoint(File("saved_stated.txt"), stage_out=True)\
                .set_stdout("out.txt")\
                .add_profiles(Namespace.PEGASUS, key="checkpoint.time", value=1)\
                .add_profiles(Namespace.PEGASUS, key="maxwalltime", value=2)

wf.add_jobs(job_add_nums)

try:
    wf.plan(submit=True)\
        .wait()\
        .analyze()\
        .statistics()
except PegasusClientError as e:
    print(e.output)       
```
